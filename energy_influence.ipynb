```python
# Format data types, separate product specifications, calculate yield
size = df["产品规格"].str.split(
    re.compile(r"[|×(]"),
    regex=True,expand=True
)[[1,2,3]].rename(columns={1:"长", 2:"宽", 3:"件/排"})
size["进砖"] = df["流量_累计进砖"]
size = size.apply(pd.to_numeric,errors="ignore")
size["产量"] = size["长"] * size["宽"] * size["件/排"] \
    * (size["进砖"]-size["进砖"].shift(1)).fillna(0)

df = pd.concat([df.drop(columns=["产品规格"]),size], axis=1)
```

```python
# Format time series
df["时间"] = pd.to_datetime(df["时间"])
df = df.set_index("时间",drop=True).apply(pd.to_numeric,errors="ignore")
df = (df-1)/1000
df
```

```python
# Count the number of output %<=15 and >15
df['开枪数'] = (df.filter(regex='输出$',axis=1)>15).apply(np.sum,axis=1)
```

```python
# Summarize every period of time
df = df.resample('8H').mean()
```

```python
# Power consumption,Gas consumption,Take the difference
df.iloc[:,-1:-3] = (df.iloc[:,-1:-3] - df.iloc[:,-1:-3].shift(1)).fillna(0)
```

```python
# Filter out the data whose energy consumption per minute changes greater than 100 kWh, linear interpolation, retain 3 decimal places
df["流量_总能耗"] = df["流量_总能耗"].where((
    df["流量_总能耗"] - df["流量_总能耗"].shift(1)
).fillna(0).abs() < 100).interpolate().round(3)
```

```python
# calculate weight
kdf = (df-df.shift(1).fillna(df)).abs()
kdf = kdf.div(kdf["流量_总能耗"],axis='index').dropna().mean()
```

```python
# Calculate cumulative absolute value
abs_df = (df-df.shift(1)).fillna(0).abs().expanding(1).sum()

# (abs_df/abs_df.max()*100).to_csv('data-abs.csv')
# .corr().dropna(how="all").dropna(axis=1,how="all").to_csv('abs-cor.csv')
# .to_csv('data-abs.csv')
```

```python
# Calculation energy consumption difference value ratio
abs_df.iloc[:,1:] = abs_df.sub(abs_df["流量_总能耗"],axis='index').abs().iloc[:,1:]
abs_df = abs_df.div(abs_df["流量_总能耗"],axis='index').drop(columns=["流量_总能耗"])
sum_df = abs_df.apply(np.sum, axis='columns')
abs_df.div(sum_df,axis='index').fillna(0).replace(1,0).to_csv('data-take.csv')
```
